# 1. Escribe una funcin que genere una poblacin inicial de cromosomas binarios de longitud fija.

```python
def rand_binary():
    p=9
    key1 = ""
    for i in range(p):
        temp = str(random.randint(0, 1))
        key1 += temp
         
    return key1
```

# 2. Escribe una funcin que mapee un cromosoma binario a valores reales.

```python
def mapeo(bin_num=''):
    num_str = bin_num
    dec_num = 0
    cont = len(num_str) - 1
    for num in range(len(num_str)) :
        if num_str[cont-num] == str(1):
            dec_num += 2**num
    return bin_num,dec_num
```

```python
mapeo(rand_binary())
```

# 3. Escribe una funcin que separe en tres partes los genes de un cromosoma binario y los mapee a valores reales como vimos en este caso de estudio.

```python
def cromosomas(binario):
    print(binario)
    crom = []
    for j in range(0,9,3):
        crom.append(mapeo(binario[j:j+3]))
    return crom
```

```python
cromosomas(rand_binary())
```
